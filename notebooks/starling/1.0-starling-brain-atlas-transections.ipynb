```python
import numpy as np
from glob import glob
from birdbrain.atlas import atlas
from birdbrain.visualization.plotting_2d import plot_transection
```

### Load data, create atlas object

```python
# where to look for the dataset
dset_dir = '../../data/processed/starling/'
# create the atlas
starling_atlas = atlas(
    species = 'starling',
    dset_dir = dset_dir,
    um_mult = 100,
    smoothing = ['Brain', 'Brainregions', 'Nuclei']
)
```

#### Plot brain regions

```python
plot_transection(starling_atlas, regions_to_plot='Brainregions', n_slice = 8);
```

### Plot nuclei

```python
plot_transection(starling_atlas, regions_to_plot='Nuclei', region_alpha = 0.5);
```
